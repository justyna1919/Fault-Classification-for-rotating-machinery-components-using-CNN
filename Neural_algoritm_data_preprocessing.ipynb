```python
import scipy.io # Do otwarcia plików '.mat'
import seaborn as sns
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
# Tworzenie pustego DataFrame z kolumnami 'DE_data' i 'fault'
df=pd.DataFrame(columns=['DE_data','fault'])

# Przeglądanie katalogu '0hp_load_48_kHz' i podkatalogów
for root, dirs, files in os.walk("0hp_load_48_kHz", topdown=False):
    for file_name in files:
        
        # Tworzenie pełnej ścieżki do pliku
        path = os.path.join(root, file_name)
        print(path)
        
        # Ładowanie danych z pliku .mat
        mat = scipy.io.loadmat(path)
        
        # Pobieranie czwartej zmiennej z pliku .mat
        key_name = list(mat.keys())[3]
        DE_data = mat.get(key_name)
        
        # Tworzenie tablicy o stałej wartości odpowiadającej nazwie pliku (bez rozszerzenia)
        fault = np.full((len(DE_data), 1), file_name[:-4])
        
        # Tworzenie tymczasowego DataFrame z danymi
        df_temp = pd.DataFrame({'DE_data':np.ravel(DE_data) , 'fault':np.ravel(fault)})
        
        # Łączenie tymczasowego DataFrame z głównym DataFrame
        df = pd.concat([df,df_temp],axis=0)
        print(df['fault'].unique())
        
# Zapisanie końcowego DataFrame do pliku CSV        
df.to_csv('0hp_all_faults.csv',index=False)
```
